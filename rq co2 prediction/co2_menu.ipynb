```python
import numpy as np
import pandas as pd
```

```python
cleaned_co2_df = pd.read_csv('cleaned_co2_df.csv')
```

```python
# Drop duplicate rows of the same dish for each day
unique_dishes_daily = cleaned_co2_df.drop_duplicates(subset=['Date', 'Dish'])

# Filter the rows where 'pcs' is 1 for each unique dish on each day
menu_df = unique_dishes_daily[unique_dishes_daily['pcs'] == 1]
menu_df = menu_df.drop(columns=['Unnamed: 0'], errors='ignore')

# Save the filtered dataframe to a new CSV file
menu_df.to_csv('menu_df.csv', index=False)
```
