## tuntidata2.xlsx

```python
df = pd.read_excel("src/data/basic_mvp_data/tuntidata2.xlsx", index_col=0)

df.head()
```

```python
df = df.replace({
    "600 Chemicum": "Chemicum",
    "610 Physicum": "Physicum",
    "620 Exactum": "Exactum"
})

df["weekday"] = df.index.dayofweek

df.head()
```

```python
grouped_df = df.groupby(
    ["Ravintola", "weekday", "Kuitin tunti"]).mean().to_dict()["Kuitti kpl"]

restaurants = ["Chemicum", "Physicum", "Exactum"]

occupancy = {
    restaurant: {
        day: [
            grouped_df.get((restaurant, day, hour), 0)
            for hour in range(24)
        ]
        for day in range(7)
    }
    for restaurant in restaurants
}
```
