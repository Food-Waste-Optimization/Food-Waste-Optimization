```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
plt.style.use('seaborn-v0_8')
plt.rcParams.update({'font.size': 8})
```

```python
path = "/home/hoanghu/Downloads/single_agent.csv"

data = pd.read_csv(path)

data.head()
```

```python
data.head()
```

```python
data = data.sort_values('dt_s')
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111)

is_first = True
x, z = 0.0, 0.0
for row in data.itertuples():
    if is_first is True:
        is_first = False
        ax.scatter(row.x_axis, row.z_axis, c='red', marker='o', s=40)
    else:
        dx, dz = row.x_axis - x, row.z_axis - z

        ax.arrow(x, z, dx, dz, head_width = 0.2)

    x, z = row.x_axis, row.z_axis

ax.set_title("Moving path of an agent", fontweight='bold', fontsize=14)
```
